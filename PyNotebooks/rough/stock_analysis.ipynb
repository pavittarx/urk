```python
from dotenv import load_dotenv;
import os;

import requests;

import pandas as pd;
import numpy as np;

load_dotenv()

ALPHA_VANTAGE_API_KEY = os.getenv('ALPHA_VANTAGE_API_KEY')

BASE_URL = 'https://www.alphavantage.co'
TICKER = 'IBM'
```

```python

weekly_data_url=f'{BASE_URL}/query?function=TIME_SERIES_WEEKLY&symbol={TICKER}&apikey={ALPHA_VANTAGE_API_KEY}'
fetched_weekly_data = requests.get(weekly_data_url).json()

# Meta Info about the Stock
print(f'{TICKER}: Meta Info')
print(fetched_weekly_data['Meta Data']);

# Raw data as provided by the API
weekly_raw = fetched_weekly_data['Weekly Time Series'];

# print(weekly_raw);

def parseto_dataframe (data):
    columns = ['Date Range', 'Open', 'High', 'Low', 'Close', 'Volume']

    df = pd.DataFrame(columns=columns)

    print("DF", df);

    for date, values in data.items():
        row = { 'Date Range': date, 
                'Open': float(values['1. open']), 
                'High': float(values['2. high']), 
                'Low': float(values['3. low']), 
                'Close':  float(values['4. close']), 
                'Volume': int(values['5. volume'])
              }
        new_row = pd.DataFrame.from_records([row], columns=columns)
        df = pd.concat([df, new_row])

    # df.set_index('Date Range', inplace=True)

    return df

weekly_data = parseto_dataframe(weekly_raw)
weekly_data
```
