- Probability is the ratio of desired outcomes to total outcomes. 

$ \text{P} = \frac{\text{desired outcomes (x)}}{\text{total outcomes (n)}}$

- Probability of all the outcomes is always equal to 1.

### Basic Terminology 

**Random Experiment**
An experiment or a process for which the outcome cannot be predicted with certainity.

**Sample Space**
The entire possible set of outcomes of a random experiment is the sample space (S) of that experiment.

**Event**
One or more outcomes of an experiment. It is a subset of sample spaces (S).




### Events

Disjoint Events do not have any common outcomes.

Non-Disjoint Events

```python
# library
import matplotlib.pyplot as plt
from matplotlib_venn import venn2

# Use the venn2 function
venn2(subsets = (1, 1, 0), set_labels = ('Event A', 'Event B'))
plt.show()


print("Disjoint Events");

venn2(subsets = (2, 2, 1), set_labels = ('Event A', 'Event B'))
plt.show()

print("Disjoint Events");
```
